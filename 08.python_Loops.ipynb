while and do while

```python
i = 0
creams = ["vanilla", "strawberry", "dark cookie"]  
while i < 3:
    i += 1
    if i == 3:
        continue
    print( creams[i] )  
```

```python
creams = ["vanilla", "strawberry", "dark cookie"]
i = 0
while True: # do part
    print(creams[i])  
    i += 1
    if i >= len(creams): 
        break
```

```python
creams = ["vanilla", "strawberry", "dark cookie"]
for cream in creams:  
    print(cream)
```
