```python
 Sets: Unordered collections of unique elements that are mutable. They are useful for membership testing and eliminating duplicate entries. Sets are defined using curly braces {} or the set() function (note that {} creates an empty dictionary).
Use Cases: Finding common items between two groups (intersection), checking for membership quickly, or removing duplicates from a list.
```

```python
sets_A={"dragonfruit", "banana", "brownberries", "orange", "watermelon", "mango"}
sets_A
```

```python

print("banana" in sets_A) # returns in boolean if matchs true ,else false.
```

```python
sets_A={"dragonfruit", "banana", "brownberries", "orange", "watermelon", "mango"}
list(sets_A) # set to list
```

```python
tuple_A=(1,2,3)
set(tuple_A) #tuple to set
```
