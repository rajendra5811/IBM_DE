```python
Dictionaries: Collections of key-value pairs that are mutable and, as of Python 3.7+, preserve the insertion order of keys. Keys must be unique and hashable (immutable types like strings, numbers, or tuples), while values can be any data type. They are defined using curly braces {}.
Use Cases: Storing and retrieving data based on a unique identifier (like a phonebook), counting the frequency of items (using collections.Counter), or managing configuration settings. 
   
```

```python
complex_numbers = {
    "natural_numbers": {
        "positive": {
            "even": {"1": "two", "2": "four", "3": "six", "4": "eight"},
            "odd": {"1": "Three", "2": "five", "3": "seven", "4": "nine"}
        },
        "negative_numbers": {"1": "-1", "2": "x<-1"}
    },
    "universal_numbers": {"1": "zero", "2": "one"}
}


result = complex_numbers["natural_numbers"]["positive"]["even"]["2"]
print(result)  # Output: four
```

```python
result = complex_numbers["natural_numbers"]["positive"]["odd"]["2"]
result
```
