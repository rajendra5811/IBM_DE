It is hard find known errors in long lines of code.Then to resolve this exception handling is introduced.
expect, try , catch are body of execption handling function.

Error: Human mistake in code or logic by developer, often caught during compilation (syntax error) or causing runtime issues.
Bug: Flaw in software causing incorrect behavior, found during testing or production. Informal term for defects impacting functionality.
Exception: Runtime event signaling abnormal condition (e.g., ZeroDivisionError, KeyError). Python raises these; handle with try-except.
Fault: Underlying system state or defect preventing correct operation, like broken logic or data corruption.
Defect: Variance between expected and actual software behavior per requirements, identified by testers.

expection handling are user defined and in-built

try: code that may cause exception
except: run the code when expection occurs
zeroDivisionerror and out of bound error
finnally is constant or default response after addressing error

```python
try:
    num = 10   
    den = 0

    result = num/den

    print(result)
except:
    print("Error: Den cannot be 0.")
```

```python
try:
    even_numbers = [2,4,6,8]
    print(even_numbers[5])   
except IndexError:
    print("Index Out of Bound.")
finally:
    print("This is similar to alert.") # finally is similar to alert
```

```python
class AgeException(Exception): #user defined expection
    pass

number = 18
try:
    input_num = int(input("Enter a number: "))
    if input_num < number:
        raise AgeException
    else:
        print("Eligible to Vote")
        
except AgeException:
    print("Exception occurred: Invalid Age")
    
```

```python
try:                                   # multiple expection
    n = 0
    res = 100 / n
    
except ZeroDivisionError:
    print("You can't divide by zero!")
    
except ValueError:
    print("Enter a valid number!")
    
else:
    print("Result is", res)
    
finally:
    print("Execution complete.")
```

```python
def set(age):
    if age < 0:
        raise ValueError("Age cannot be negative.")
    print(f"Age set to {age}")

try:
    set(-5)
except ValueError as e:
    print(e)
```

```python
def validate_salary(salary):
    try:
        if not 5000 < salary < 15000:
            raise ValueError(f"Salary {salary} is not in (5000, 15000) range") # f" is like %d to get salary'value.
        return "Valid"
    except ValueError as e:
        return str(e)

salary = int(input("Enter salary amount: "))
result = validate_salary(salary)
print(result)
```
