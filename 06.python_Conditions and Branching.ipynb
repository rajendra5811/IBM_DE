elif is condition with if inside of else 

```python
grade = 85
if grade >= 90: # if else ladder with elif
    result = "A"
elif grade >= 80:
    result = "B"
elif grade >= 70:
    result = "C"
else:
    result = "F"
print(result)  # "B"
```

```python
status = 404   #switch case
match status:
    case 200:
        msg = "OK"
    case 404:
        msg = "Not Found"
    case _:
        msg = "Unknown"
print(msg)  # "Not Found"
```
