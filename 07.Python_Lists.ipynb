List comprehension provides concise syntax to create lists by applying expressions to iterables, replacing traditional for loops.

```python
newlist = [x for x in list_a if x != "sham"]  
newlist # expect sham all returned
```

```python
newlist.sort()
newlist  #sort(reverse = True)
```
